```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrames
data1 = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['John', 'Alice', 'Bob', 'Emma', 'Mike'],
    'Height': [175, 162, 180, 168, 173]
}

data2 = {
    'ID': [6, 7, 8, 9, 10],
    'Name': ['Sarah', 'David', 'Sophie', 'Tom', 'Emily'],
    'Height': [155, 185, 170, 165, 170]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Combine both datasets
combined_df = pd.concat([df1, df2])

# Find lowest and highest values in the combined "Height" column
lowest_height = combined_df['Height'].min()
highest_height = combined_df['Height'].max()

# Manually create 10 bins
bins = [lowest_height + i * ((highest_height - lowest_height) / 10) for i in range(11)]

# Group heights into 10 bins for each DataFrame
df1['Height Group'] = pd.cut(df1['Height'], bins=bins, include_lowest=True)
df2['Height Group'] = pd.cut(df2['Height'], bins=bins, include_lowest=True)

# Count the number of people in each bin for each dataset
height_groups_df1 = df1.groupby('Height Group').size()
height_groups_df2 = df2.groupby('Height Group').size()

# Create a bar chart with bars for each dataset, colored differently
fig, ax = plt.subplots()
index = range(len(height_groups_df1))
bar_width = 0.35

bar1 = ax.bar(index, height_groups_df1, bar_width, label='Dataset 1', color='blue')
bar2 = ax.bar([i + bar_width for i in index], height_groups_df2, bar_width, label='Dataset 2', color='orange')

ax.set_xlabel('Height Groups')
ax.set_ylabel('Number of People')
ax.set_title('Number of People in Height Groups')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels([f'{bins[i]:.0f}-{bins[i+1]:.0f}' for i in range(len(bins)-1)], rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```
